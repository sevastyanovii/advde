```python
# Задание 1
# Скачайте с сайта grouplens.org...movielens
# датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

import pandas as pd
data = pd.read_csv('data/ratings.csv')
# data.head()
data[data['rating'] == 5].groupby(['movieId'])['movieId'].count() \
    .reset_index(name='count')\
    .sort_values(['count'], ascending=False) \
    .head()
```

```python
# Задание 2
# По данным файла power.csv посчитайте суммарное потребление стран 
# Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. 
# Не учитывайте в расчетах отрицательные значения quantiy.

country_dict = {
    'Latvia': '1', 
    'Lithuania': '1', 
    'Estonia': '1'
}

import pandas as pd
data = pd.read_csv('data/power.csv')
data['region'] = data['country'].map(country_dict).fillna('0')
a = data.query('region == "1" and quantity > 0 and year >= 2005 \
           and year <= 2010 and category == [4, 12, 21]')
a[['country','year','quantity','category']]
```
